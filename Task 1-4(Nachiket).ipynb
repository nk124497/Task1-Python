**4) Python Program to Swap the First and Last Value of a List.**

```python
list
```

```python
def swap(list): 
    size = len(list) 
    temp = list[0] 
    list[0] = list[size - 1] 
    list[size - 1] = temp 
      
    return list 
```

```python
print(swap(list))
```
